```python
from pprint import pprint
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from gensim.test.utils import datapath, get_tmpfile
from gensim.models import KeyedVectors
from gensim.scripts.glove2word2vec import glove2word2vec
```

```python
glove_file = datapath('/Users/zhoujiagen/temp/nlp/glove.6B.300d.txt')
tmp_file = get_tmpfile("/Users/zhoujiagen/temp/nlp/glove.6B.300d_word2vec.txt")

_ = glove2word2vec(glove_file, tmp_file)
model = KeyedVectors.load_word2vec_format(tmp_file)
```

```python
dir(model)
```

```python
man_v = model.wv['man']
woman_v = model.wv['woman']
king_v = model.wv['king']
# king-man+woman
pprint(king_v - man_v + woman_v)
```

```python
words = ['queen', 'woman', 'girl', 'boy', 'man', 'king', 'water']
words_v = []
for word in words:
    words_v.append(model.wv[word]) 
```

```python
#import numpy as np
fig, ax = plt.subplots()
#ax.set_xticks(np.arange(300), minor=False)
ax.set_yticklabels(words, minor=False)
ax.pcolor(words_v, cmap=plt.cm.Blues)
fig.set_size_inches(40, 10)
plt.show()
```
