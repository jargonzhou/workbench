```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data

import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
DATA_DIR = '/tmp/mnist' # data to save locally
mnist = input_data.read_data_sets(DATA_DIR, one_hot=True)
```

```python
first_image = mnist.test.images[0]
first_image = np.array(first_image, dtype='float')
pixels = first_image.reshape((28, 28))
plt.imshow(pixels, cmap='gray')
plt.show()
```

```python
#for i in range(3):
#    print(mnist.test.labels[i])
#    print(np.argmax(mnist.test.labels[i]))

print("Total size: ", len(mnist.test.images))

plt.gray()
for i in range(10):
    plt.title("Label: {}".format(str(np.argmax(mnist.test.labels[i]))))
    plt.imshow(mnist.test.images[i].reshape(28,28))
    plt.show()
```
