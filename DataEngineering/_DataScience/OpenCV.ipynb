```python
plt.figure(figsize = (60, 40))
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.axis("off")
plt.show()
```

# 阈值

```python
ret,img = cv2.threshold(img,160,255,cv2.THRESH_BINARY)

plt.figure(figsize = (60, 40))
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.axis("off")
plt.show()
```

# 形态学

```python
#kernel = np.ones((1,1),np.uint8)
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(1,1))
processed_img = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
# opening: MORPH_OPEN
# gradient: MORPH_GRADIENT

plt.figure(figsize = (60, 40))
plt.imshow(processed_img, cmap='gray', vmin=0, vmax=255)
plt.axis("off")
plt.show()
```
