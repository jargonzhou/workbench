# CNN

- each unit is connected to a (typically small) number of nearby units in the previous layer
- all units are connected to the previous layer in the same way, with the exact same weights and structure.



motivations:

- neuroscientific inspiration behind the model

The current understanding in the neuroscientific community is that as visual processing proceeds, information is integrated from increasingly wider parts of the input, and this is done hierarchically.

Each convolutional layer looks at an **increasingly larger part** of the image as we go deeper into the network

- insight into the nature of images

When looking for an object in an image, say the face of a cat, we would typically want to be able to detect it regardless of its position in the image. This reflects the property of natural images that the same content may be found in different locations of an image.

**invariances** of this sort can also be expected with respect to (small) rotations, changing lighting conditions, etc.

- learning theory

as a regularization mechanism

In machine learning and statistics, **regularization** is mostly used to refer to the restriction of an optimization problem by imposing a penalty on the complexity of the solution, in the attempt to prevent overfitting to the given examples.

## Convolution


```
tf.nn.conv2d(x, W, strides=[1,1,1,1], padding='SAME')
```

- shape of x: \[None, 28, 28, 1\]

image count/image width/image height/image color channel

- shape of W: \[5, 5, 1, 32\]

convolved window(5x5x1)/the number of feature maps

- strides

control the spatial movement of W across x

- padding

the borders of x are padded
