```python
import numpy as np
import pandas as pd

# from pandas import Series, DataFrame


def print_df(df: pd.DataFrame):
    from IPython.core.display import display_html, HTML
    display_html(HTML(df.to_html()))
```

# Object creation

```python
# Series: a one-dimensional labeled array holding data of any type
s = pd.Series([1, 3, 5, np.nan, 6, 8])
s
```

```python
# DataFrame: a two-dimensional data structure that holds data like a two-dimension array or a table with rows and columns
dates = pd.date_range("20130101", periods=6)
dates
```

```python
# NumPy array with a DatetimeIndex
df = pd.DataFrame(np.random.randn(6, 4), index=dates, columns=list("ABCD"))
df
```

```python
# dict of objects: key are columns, values are column values
df2 = pd.DataFrame({
    "A": 1.0,
    "B": pd.Timestamp("20130102"),
    "C": pd.Series(1, index=list(range(4)), dtype='float32'),
    "D": np.array([3] * 4, dtype="int32"),
    "E": pd.Categorical(["test", "train", "test", "train"]),
    "F": "foo"
})
df2
```

## Selection by position

```python
# DataFrame.iloc()
# DataFrame.iat()

# 4th row
print_df(df)
df.iloc[3]
```

```python
# integer slice
print_df(df)
df.iloc[3:5, 0:2]
```

```python
# list of integer position locations
print_df(df)
df.iloc[[1, 2, 4], [0, 2]]
```

```python
# slicing rows
print_df(df)
df.iloc[1:3, :]
```

```python
# slicing columns
print_df(df)
df.iloc[:, 1:3]
```

```python
# cell value
print_df(df)
df.iloc[1, 1], df.iat[1, 1]
```

## Boolean indexing

```python
# where df.A > 0
df[df["A"] > 0]
```

```python
# selecting values from DataFrame where boolean condition is met
df[df > 0]
```

```python
# isin(): filtering
df2 = df.copy()
df2["E"] = ["one", "one", "two", "three", "four", "three"]
print_df(df2)

df2[df2["E"].isin(["two", "four"])]
```

## Setting

```python
print_df(df)
# set a new column: align the data by indexes
df["F"] = pd.Series([1, 2, 3, 4, 5, 6],
                    index=pd.date_range("20130102", periods=6))
print_df(df)
```

```python
# set values by label
print_df(df)
df.at[dates[0], "A"] = 0
print_df(df)
```

```python
# set value by position
print_df(df)
df.iat[0, 1] = 0
print_df(df)
```

```python
# set by assign a NumPy array
print_df(df)
df.loc[:, "D"] = np.array([5] * len(df))
df
```

```python
df2 = df.copy()
print_df(df2)
# set with where
df2[df2 > 0] = -df
df2
```
