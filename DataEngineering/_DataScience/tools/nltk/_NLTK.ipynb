```python
from nltk.corpus import reuters

fileids = reuters.fileids()
test_ids = [int(fileid[5:]) for fileid in fileids if fileid.startswith('test')]
training_ids = [int(fileid[9:]) for fileid in fileids if fileid.startswith('training')]
print("{}, {}".format(min(test_ids), max(test_ids)))
print("{}, {}".format(min(training_ids), max(training_ids)))
```

```python
# 在Jupyter notebook中使用Matplotlib
%matplotlib inline
```

```python
min_test_fileid = min(test_ids)
cdf = nltk.ConditionalFreqDist(
        (target, fileid[5:])
        for fileid in reuters.fileids()[:len(fileids) // 10] if fileid.startswith('test')
        for w in reuters.words(fileid)
        for target in ['china', 'american']
        if w.lower().startswith(target))

cdf.plot()
```
